Define Target Variable (y) and Feature Variables (X)

```python
y = data['Item_Outlet_Sales']
```

```python
y.shape
```

```python
y
```

```python
X = data[['Item_Weight', 'Item_Fat_Content', 'Item_Visibility', 'Item_Type',
          'Item_MRP', 'Outlet_Identifier', 'Outlet_Establishment_Year',
          'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type']]
```

```python
x = data.drop(['Item_Identifier', 'Item_Outlet_Sales'], axis=1)
```

```python
X.shape
```

```python
X
```

Get x variable standarized

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_std = data[['Item_Weight', 'Item_Visibility', 'Item_MRP', 'Outlet_Establishment_Year' ]]
```

```python
X_std = sc.fit_transform(X_std)
```

```python
X_std
```

```python
X[['Item_Weight', 'Item_Visibility', 'Item_MRP', 'Outlet_Establishment_Year' ]] = pd.DataFrame(X_std, columns= [['Item_Weight', 'Item_Visibility', 'Item_MRP', 'Outlet_Establishment_Year']])
```

```python
X
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.1, random_state=2529)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

Modeling Train


```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(random_state=2529)
```

```python
rfr.fit(X_train, y_train)
```

```python
y_predict = rfr.predict(X_test)
```

```python
y_predict.shape
```

```python
y_predict
```

Model Evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mean_squared_error(y_test, y_predict)
```

```python
mean_absolute_error(y_test, y_predict)
```

```python
r2_score(y_test, y_predict)
```

Get visualization Actual vs Predict




```python
import matplotlib.pyplot as plt
plt.scatter(y_test, y_predict)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predict Price")
plt.show()
```
